```python
# Dependencies and Setup
import pandas as pd
from pymongo import MongoClient
from pprint import pprint

# File to Load
file1 = "project3.csv"

# Read the File into Pandas DataFrames
projectDF= pd.read_csv(file1) 
projectDF.head(20)
```

```python
#create a new dataframe
reduced_DF = projectDF[[
          "Diabetes_binary",
          "HighBP",
          "HighChol",
          "Smoker",
          "Stroke",
          "HeartDiseaseorAttack",
]]


reduced_DF.head()
```

```python
# convert the dataframe to a list of dictionaries to be prepared to import into MongoDB
reduced_data= reduced_DF.to_dict('records')


```

```python
# create an instance of MongoClient
mongo = MongoClient(port=27017)
# make a reference to the new database named "project3" with the collection named test_project3
db = mongo.project3
collection = db.test_project3
```

```python
# use insert_many() to insert data from the extracted DataFrame info
collection.insert_many(reduced_data)
```
